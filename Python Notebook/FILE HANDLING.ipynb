```python
#  writelines - it can pass list and string

Hobbies = ["Scrolling_Feeds", " Sports", " Travel"]

file_3 = open("List_File","w")

file_3.writelines(Hobbies)

file_3.close()
```

```python
# suppose we can pass list in write() , TypeError : write() argument must be str, not list

file_3 = open("List_File","w")

file_3.write(Hobbies)

file_3.close()
```

```python
# append (a) -> add an item from existing files

file_2 = open("write_file","a")

file_2.write("\nI have added extra item using append mode")

file_2.close()
```

```python
# creating a file (using append(a) mode)

file_4 = open("Append_File","a")

file_4.write("Newly Added File in Append Mode")

file_4.close()
```

#                        READING  A  FILE  CONCEPT


```python
# Creating a New File 

file_5 = open("New_File","x")
```

```python
# Writing a Content to the File

file_5 = open("New_File","w")

file_5.write("Hi Learners\n")
file_5.write("Hope you are Beginners\n")
file_5.write("To Learn Data Science Course\n")
file_5.write("Welcome to GUVI")

file_5.close()
```

```python
# Reading a file (r)

file_5 = open("New_File","r")

file_5.read()   # o/p : no elements present in o/p

file_5.close() 
```

```python
# Reading a file (r)

file_5 = open("New_File","r")

print(file_5.read())  # print(file_5.read())

file_5.close()
```

```python
# readlines()  - > it returns all items in a list format

file_5 = open("New_File","r")

print(file_5.readlines())

file_5.close()
```

```python
# readline() -> it reads 1st value of the file

file_5 = open("New_File","r")

print(file_5.readline())  # while giving print() -> it gives space and
                          #                         pointers come into the next line

print(file_5.readline())
print(file_5.readline())

file_5.close()
```

```python
# readline() with end parameter  

file_5 = open("New_File","r")

print(file_5.readline(),end="")  # end =""
print(file_5.readline(),end="")
print(file_5.readline())

file_5.close()
```

```python
# readline() with numerical parameter  -> fetching the characters based on numbers

file_5 = open("New_File","r")

print(file_5.readline(5))  # fetch 1st 5 characters
print(file_5.readline(6))  # fetch another 6 characters

file_5.close()
```

```python
# Looping Through

file_5 = open("New_File","r")

for data in file_5:
    print(data)
    
file_5.close()    
```

```python
# Examples :

file_5 = open("New_File","r")
file_6 = open("Write_newFile","w")

for data in file_5:
    file_6.write(data)
    
file_5.close()
file_6.close()
```


#              Reading a Binary File(Image file)

###   download any pic from the internet :  

```python
# Examples :

f1 = open("jupyter_Image.jpg","r")  # we can give 'rb' mode while reading binary image file. 

for i in f1:
    print(i)     # UnicodeDecodeError  
    
f1.close()    
```

```python
f1 = open("jupyter_Image.jpg","rb")  # we can give 'rb' mode while reading binary image file. 

for i in f1:
    print(i)       
    
f1.close()  
```

```python
# Addding same image in different files (using 'wb' mode)

f1 = open("jupyter_Image.jpg","rb")

f2 = open("New_Jupyter_Image.jpg","wb")  # wb -write binary

for i in f1:
    f2.write(i)
    
f1.close()
f2.close()
```
