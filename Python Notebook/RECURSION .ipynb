#                              RECURSION

       -  A function that calls itself.
       
       -  Python has a default maximum recursion depth: 1000.
       
##  why do we use recursive function??

       -  break 'complicated' problems down into 'simple' problems 
       -  which are easier to solve.
       
###  Examples :

       -  Factorial
       -  Fibonacci series

```python
# how to find or get "default recursive depth limit"

import sys

print(sys.getrecursionlimit())
```

```python
# how to set the depth limit for the recursive functions

import sys

sys.setrecursionlimit(1000)

print(sys.getrecursionlimit())
```

```python
# code for " default maximum recursion depth of 1000. "

# EXAMPLES :

i = 0
def welcome():
    
    global i
    i = i+1
    
    print("Welcome = ",i)
    
    welcome()  # recursive function - > function that calls itself .
    
welcome()
```

```python
# Factorial of a Number [using RECURSIVE Function]

# code:
def fact(n):
    
    if(n == 0):
        return 1
    
    return n * fact(n-1)

factorial = fact(5)
print("Factorial = ",factorial)
```
