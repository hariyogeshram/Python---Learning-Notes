```python
# Import the Libraries

import mysql.connector
```

```python
# connect the mysql database

my_connection = mysql.connector.Connect(host = "localhost", port="3306", user="root", password="", database="bizcardx")

# cursor object
cur = my_connection.cursor()

# select query

query = "SELECT * FROM business_card;"

# Execute the Query
cur.execute(query)

# Fetch all databases to clear any unread results
records = cur.fetchall()

# print("No.of Records present in the Business card : ", cur.rowcount())

print("-----------------   Business Card Details   ---------------------------")

print("\n")

# Check the list of records in the databases
for row in records:
    print("Company Name : ", row[0])
    print("Card Holder Name : ", row[1])
    print("Designation : ", row[2])
    print("Website URL : ", row[5])
    print("City : ", row[7])
    print("\n")


# you can close the cursor if you're done with it
cur.close()

# Close the connection as well
my_connection.close()
```

```python
# import mysql.connector

# Using Function

def fetch_business_card_details(host, port, user, password, database):

    try:

        # Establish the MySQL connection
        my_connection = mysql.connector.Connect(host=host, port=port, user=user, password=password, database=database)
        
        # Create a cursor object
        cur = my_connection.cursor()
        
        # Define the query
        query = "SELECT * FROM business_card;"
        
        # Execute the query
        cur.execute(query)
        
        # Fetch all the records
        records = cur.fetchall()
        
        # Check if there are records
        if records:

            print("-----------------   Business Card Details   ---------------------------")
            print("\n")
            
            # Loop through the records and print the details
            for row in records:

                print("Company Name : ", row[0])
                print("Card Holder Name : ", row[1])
                print("Designation : ", row[2])
                print("Website URL : ", row[5])
                print("City : ", row[7])
                print("\n")

        else:

            print("No records found.")
        
        # Close the cursor
        cur.close()
        
        # Close the connection
        my_connection.close()

    except mysql.connector.Error as err:

        print(f"Error: {err}")

    finally:

        if my_connection.is_connected():

            my_connection.close()

            print("MySQL connection is closed")

# Function call with parameters:

fetch_business_card_details(host="localhost", port="3306", user="root", password="", database="bizcardx")
```
