```python
# operator()

              #  The operator module contains functions that correspond to standard mathematical operations
              #  like addition, subtraction, multiplication, division, etc.

              #  These functions are useful in functional programming.

              # import operator
```

```python
# Examples :

import operator

# Example of using operator for "multiplication"

result = operator.mul(2, 3)

print("multiplication = ", result)

# Example of using operator for "addition"

result = operator.add(10, 5)

print("addition = ", result)

# Example of using operator for "Subtraction"

result = operator.sub(10, 5)

print("Subtraction = ", result)

# Example of using operator for "true division"

result = operator.truediv(10, 5)

print("true division = ", result)

# Example of using operator for "floor division"

result = operator.floordiv(10, 5)

print("floor division = ", result)
```

```python
# Using reduce

# We can perform operator methods

from functools import reduce

import operator

lists = [2,4,6,8,9,5,3]

print("Sum of list elements = ", end=" ")

print(reduce(operator.add, lists))

print("Product of list elements = ", end=" ")

print(reduce(operator.mul, lists))

```

```python
# Definition:

              #  itertools.accumulate() function returns a cumulative sum (or another binary operation result) of the values in an iterable.
              #  By default, it performs cumulative addition,
              #  but you can specify other operations using the operator module or a custom function.
```

```python
# Example :

import itertools

# Cumulative sum of numbers:

numbers = [1, 2, 3, 4]

result = list(itertools.accumulate(numbers))

print(result)  # Output: [1, 3, 6, 10]
```

```python
import itertools
import operator

# Cumulative product of numbers using operator.mul

numbers = [1, 2, 3, 4]

result = list(itertools.accumulate(numbers, operator.mul))

print(result)  # Output: [1, 2, 6, 24]
```
