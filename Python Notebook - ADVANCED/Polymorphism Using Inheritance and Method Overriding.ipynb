```python
# Examples:

# Polymorphism Using "Inheritance" and "Method Overriding"

class Vehicle():

  def __init__(self, name, color, price):

    self.name = name
    self.color = color
    self.price = price

  def display(self):

    print("*******************************    Vehicle Details   ******************************** \n\n")

    print("\n Vehicle Name = ", self.name)
    print("\n Vehicle Color = ", self.color)
    print("\n Vehicle Price = ", self.price)

  def speed(self):

    print("\n Max Speed of Scooty is 120")

  def gears(self):

    print("\n Scooty has NO gears")


class Bike(Vehicle):

  def speed(self):

    print("\n Max Speed of Bike is 160")

  def gears(self):

    print("\n Bike has 4 gears")


class Car(Vehicle):

  def speed(self):

    print("\n Max Speed of Car is 240")

  def gears(self):

    print("\n Car has 6 gears")



b = Bike("R15", "Black", "2 Lakhs")

b.display()

b.speed()

b.gears()

print('\n')

c = Car("Renault Triber", "Blue", "8.5 Lakhs")

c.display()

c.speed()

c.gears()

```

```python
# Examples

# Using Polymorphism - Classes , Functions and objects

class Person_1():

  def name(self):

    print("Name = Yogesh")

  def age(self):

    print("Age = 24")

  def gender(self):

    print("Gender = Male")


class Person_2():

  def name(self):

    print("Name = Kanagavalli")

  def age(self):

    print("Age = 56")

  def gender(self):

    print("Gender = Female")


def function_call(obj):

  obj.name()
  obj.age()
  obj.gender()

# Obj creation

p1 = Person_1()
p2 = Person_2()

# Function call inside we can pass objects

function_call(p1)
print("\n")
function_call(p2)
```
