```python
# Constructor:

# Initialize(Assign Values) to the data members of the class
# when an object of the class is created.
```

```python
# Types of Constructor:

# 1) Default constructor

# 2) Parameterized Constructor
```

```python
# Example - 1 :   Default Constructor

class Person():

  def __init__(self):

    name = "Yogesh"
    age = 24
    gender = "Male"

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)


obj = Person()
```

```python
# Example - 2 :   Default Constructor

# more than 1 constructor giving , overriding can be achieved , so latest one will be returned

class Person():

  def __init__(self):

    name = "Yogesh"
    age = 24
    gender = "Male"

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)

  def __init__(self):

    name = "Ram"
    age = 22
    gender = "Male"

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)


obj1 = Person()

print("\n")

obj2 = Person()
```

```python
# Example - 1 : Parameterized Constructor

class Person():

  def __init__(self, name, age, gender):

    self.name = name
    self.age = age
    self.gender = gender

  def print_person_details(self):

    print("Name : ", self.name)
    print("Age : ", self.age)
    print("Gender : ", self.gender)



# obj creation

obj1 = Person("Ganesh", 23, "Male")

obj1 = Person("Kavya", 21, "Female")

obj1.print_person_details()
```

```python
# Example - 2 : Parameterized Constructor

class Triangle():

  def __init__(self, b, h):

    self.base = b
    self.height = h

  def area_of_Triangle(self):

    self.area = 0.5 * self.base * self.height      # area = (1/2)*bh

    print("Area of a Triangle = ", self.area)



# object creation

obj = Triangle(b=12, h=30)

obj.area_of_Triangle()
```

```python
# Constructor with " Return type"  =>  It should always returns 'None', not 'int'

# Examples:

class Addition():

  def __init__(self, a, b):

    self.a = a
    self.b = b

    self.c = self.a + self.b

    return self.c



# obj creation

add = Addition(10, 20)
```

```python
#                                                                     Destructor  (  def __del__(self)  )

# It is called => When an object is destroyed

# Main Uses :

#              1) De-allocate Memory
#              2) Clean up Resources
```

```python
# Example - Destructor

class Person():

  def __init__(self, name, age):

    self.name = name
    self.age = age

  def print_details(self):

    print("Name : ", self.name)
    print("Age: ", self.age)


  # destructor

  def __del__(self):

    print("Destructor is called!.....Clean up the Resources.")



# obj creation

obj = Person("Gopi", 29)

obj.print_details()

print("\n")


# calling the destructor

del obj

# obj.print_details()   # after destructing, when calling the method again, it will return an NameError : name 'objname' is not defined
```
