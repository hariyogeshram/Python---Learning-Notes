```python
# Passing "function" as a (Parameter)
```

```python
# Example - 1:

def add(a, b):

  return a + b


def display(func, a, b):

  return func(a, b)


result = display(add, 8, 9)

print("Addition of 2 numbers = ", result)
```

```python
# Example - 2:

def upper(string):

  return string.upper()

def lower(string):

  return string.lower()


def title(string):

  return string.title()


def capitalize(string):

  return string.capitalize()


def string(func):

  message = func("welcome to python")

  print(message)


string(upper)

string(lower)

string(title)

string(capitalize)

```
