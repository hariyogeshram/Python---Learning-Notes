```python
# Encapsulation

# 'bundling or Wrapping' of data (attributes) and methods (functions)
# that operate on the 'data' into a (single unit or class).

# Example : Capsule
```

```python
# Encapsulation helps to 'protect' an (objectâ€™s internal state )
# And only allows it to be modified in a controlled way via (getter and setter methods).
```

```python
# Key Types of Encapsulation:

# Public members:      Accessible from anywhere.                                  EXAMPLE :    self.brand = brand  # Public attribute

# Protected members:   Accessible only within the class and its subclasses.       EXAMPLE :    self._brand = brand  # Protected attribute  (_)  => SINGLE underscore

# Private members:     Accessible only within the class itself                    EXAMPLE :    self.__brand = brand  # Private attribute   (__) => Double Underscore
```

```python
# Examples :

# 1) # Public members:      Accessible from anywhere.

class School():

  def details(self, name, city):

    self.name = name
    self.city = city

    print("School Name : ", self.name)


# object Creation

obj = School()

obj.details("AVMS SCHOOL", "Rajapalayam")

print("School City : ", obj.city)     # Public Members


```

```python
# Private members:     Accessible only within the class itself  (__)

# Examples

class School():

  def details(self, school_name, city):

    self.school_name = school_name
    self.city = city

    print("School Name : ", self.school_name)
    print("School Location : ", self.city)

  # Private methods
  def __student1(self):

    print("Student Name : Hari")
    print("Class : 10th STD\n")

  #  Public method
  def student2(self):

    # We can access the above private methods to invoke

    self.__student1()

    print("Student Name : Yogesh")
    print("Class : 12th STD")



# Object

obj = School()

obj.details("AVMS", "Rajapalayam")

print("\n")

obj.student2()
```

```python
# Name Mangling

# Examples ( same above code , but little differences)

# Private members:     Accessible only within the class itself  (__)

# Examples

class School():

  def details(self, school_name, city):

    self.school_name = school_name
    self.city = city

    print("School Name : ", self.school_name)
    print("School Location : ", self.city)

  # Private methods
  def __student1(self):

    print("Student Name : Hari")
    print("Class : 10th STD\n")

  #  Private method
  def __student2(self):

    print("Student Name : Yogesh")
    print("Class : 12th STD")



# Object

obj = School()

obj.details("AVMS", "Rajapalayam")

print("\n")

# obj.student2()

# Name Mangling =>   syntax :   objname._className__MethodName()

obj._School__student1()

obj._School__student2()


```

```python
# 3) Protected members:   Accessible only within the class and its subclasses.

# Examples:

class Base():

  def __init__(self):

    # Protected variables
    self._name = "Yogesh"


class Derived(Base):

  def __init__(self):

    super().__init__()

    print(" Calling from Derived Class : ", self._name)

    self._name = "Ram"

    print("\n Modified Name : ", self._name)



obj1 = Derived()

obj2 = Base()

print("\n Accessing Protecting Member of object 1 = ", obj1._name)

print("\n")

print(" Accessing Protecting Member of object 2 = ", obj2._name)
```

```python
# setter and getter method in detail

# Examples

class Set_And_Get_Age():

  def __init__(self):

    self._age = 0


  # getter method

  @property

  def age(self):

    print("Getter Method is Called")

    return self._age


  # setter Method :  syntax : @methodName.setter

  @age.setter

  def age(self, a):

    if a < 18:

      print("You are Under 18!..")

    print("Setter Method is called")

    self._age = a



# object creation

obj = Set_And_Get_Age()

obj.age = 25  # setter method calling

print(obj.age)  # calling getter method

```
