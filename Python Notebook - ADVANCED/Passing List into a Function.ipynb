```python
#  Checking how many "even and odd numbers" present in the given list

# define a function

def check_how_many_even_odd_nos_present(given_list):

  odd = 0
  even = 0

  for i in given_list:

    if i % 2 == 0:

      even = even + 1

    else:

      odd = odd + 1

  return even, odd


# input

given_list = [8, 18, 45, 93, 7, 12, 55, 17, 33, 63, 19, 10, 48, 3, 54]

# Function call:

even, odd = check_how_many_even_odd_nos_present(given_list)


print("Total Number of elements present in the List = ", len(given_list))

print("\n")

print("No. of EVEN Elements present in the List = ", even)

print("No. of ODD Elements present in the List = ", odd)


```
