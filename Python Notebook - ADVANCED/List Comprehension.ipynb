```python
# List Comprehension

# Definition:

              # It is a 'concise way' to (create -> "lists").
              # It allows you to "generate a new list"
              # by (applying an expression) to "each item" in an "existing iterable" (like lists, strings, or tuples) in a single line of code.

              # It can (optionally) "include a condition" to (filter -> "items").
```

```python
# Syntax :

#                var_name = [ expression for item in iterable if condition ]



# expression:  Defines how the element will be transformed.

# item:        Represents each element in the iterable.

# iterable:    The collection being iterated (e.g., list, range).

# condition:   (Optional) A filtering condition.
```

```python
# Example - 1 : [ list of "Squares"]

result = [ x**2 for x in range(6)]

print(result)
```

```python
# Example - 2 : [ List of EVEN Numbers]

result = [ x for x in range(11) if x % 2 == 0]

print("Even Numbered List = ", result)
```

```python
# Example - 3 : [retrieve only "list of digits" in the sequence]

digits = "hariyogeshram09052000"

result =  [ x for x in digits if x.isdigit()]

print(result)
```

```python
# Example - 4 : [ Convert Lowercase to Uppercase]

sequence = ["happy", "morning", "welcome", "to", "python"]

result = [x.upper() for x in sequence]

print(result)
```

```python
# Example - 5 :  [ Using "Nested Loops": Multiplying elements from two lists ]

list_1 = [1, 2, 3]

list_2 = [4, 5, 6]

products = [ x * y for x in list_1 for y in list_2]

print(products)
```

```python
# Advantages:

# Concise and readable.

# Efficient in terms of code structure and performance for small operations.
```

```python
# Disadvantages:

# Can become hard to read with complex logic.

# Not ideal for very large operations due to memory constraints.
```
