```python
var = "hello"

def greet():

  global var

  var = "Bye Bye!.."

  print("Inside Function variables : ", var)


greet()

print("Outside Function variables : ", var)
```

```python
# Global and Local variable having same name

var = "hello"

def greet():

  # global var

  var = "Bye Bye!.."

  print("Inside Function variables : ", var)

  print(globals()['var'])


greet()

print("Outside Function variables : ", var)
```
