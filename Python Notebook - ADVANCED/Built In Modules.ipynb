```python
# Modules Definition:

#                    A module in Python is a file containing Python code (functions, classes, variables, etc.)
#                    That can be imported and reused in other Python programs.
#                    It helps in organizing code into smaller, manageable components.
```

```python
# Main Uses:


# 1) Code Reusability:

#                         -> Modules help avoid rewriting code
#                         -> by allowing you to use functions and classes across multiple programs.

# 2) Code Organization:

#                         -> By splitting code into different modules,
#                         -> the program becomes more structured and manageable.

# 3) Namespace Management:

#                          -> Modules create their own namespaces,
#                          -> preventing name conflicts between variables or functions in different parts of a program.
```

```python
# Built in Modules

import math

n = int(input("Enter the Number : "))

result = math.sqrt(n)

print(f"Square root of {n} = {result}.")
```
