```python
# class

# syntax:

#         class classname():

# eg:

class person():

    a = 10     # attributes


# obj creation:

# syntax :

#             objectname = classname()

obj = person()

print(obj.a)

```

```python
# Oops - Methods

class Text():

  def greet(self, name):

    return f"Welcome to {name} course!..Happy Learning."


# obj creation:

name = input("Enter the Course Name : ")

obj = Text()

obj.greet(name)
```

```python
# Another way of object creation using by calling class name with method name

class Text():

  def greet(self, name):

    return f"Welcome to {name} course!..Happy Learning."


# obj creation:

name = input("Enter the Course Name : ")

obj = Text()

# Calling the object

# className.methodName(object)

Text.greet(obj, name)
```

```python
# Another Example : self parameter

class Person():

  def get_details(self, name, age, gender):

    self.name = name
    self.age = age
    self.gender = gender

  def print_details(self):

    print("Name : ", self.name)
    print("Age : ", self.age)
    print("Gender : ", self.gender)


# obj creation

obj = Person()

# Function call

obj.get_details("Yogesh", 24, "Male")

obj.print_details()

```
