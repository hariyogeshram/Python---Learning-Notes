```python
# Lambda function (Anonymous Function)


# Definition :

                  # It can take (any number of arguments) but have "only one expression".

                  # They are typically used for "short, throwaway functions or one-liners "

                  # that "donâ€™t require" a( formal function definition) using "def".
```

```python
# Syntax:

#            lambda arguments: expression
```

```python
# Example ( Using Normal function --> def)

def add(a, b):

  return a+b

result = add(45, 93)

print("Addition Using 'Normal Function' = ", result)
```

```python
# Example ( Using Lambda function )

# Addition of 2 numbers:

add = lambda a, b : a + b

print("Addition Using Lambda Function = ", add(93, 45))
```

```python
# Lambda - User Defined Function

# Examples:

add = lambda *args : sum(args)

print("Sum of all Arguments = ", add(10, 20, 30, 40, 50, 60))
```

```python
# another way of declaring lambda function followed by parameter

(lambda a, b : a ** b)(10, 3)
```

```python
# Lambda - User Defined Function [ square of each number in the list]

list_1 = [4, 9, 5, 8, 3, 6]

list_2 = []

for i in list_1:

  temp = lambda i : i ** 2

  list_2.append(temp(i))


print(list_2)
```

```python
# Lambda - User Defined Function:

def a(x):

  return(lambda y : x + y)


result = a(5)

print(result(8))
```

```python
# Lambda with Higher Order Function

# A function in Python with another function as an "argument" or "Returns a function" as an "Output"

higher_order = lambda x, function : x + function(x)

higher_order(20, lambda x : x ** 2)     # 20 + 400 => 420
```

```python
# 1) map():

          # The map() function applies a (given function to all items) in an "iterable" (like a list, tuple, etc.)
          # "Returns" a (new iterable)


# Syntax :     map(function, iterable)
```

```python
# Examples => only with map()

# Check odd or even numbers in the list

def odd_even(num):   # function

  if num % 2 == 0:

    return "Even"

  else:

    return "Odd"

# iterable

numbers = [7, 18, 10, 45, 8, 93, 33]

# Syntax :  list(map(function, iterable))

result = list(map(odd_even, numbers))

print("Check Odd or Even numbers in the given list : \n\n", result)
```

```python
# Examples => only with map() with lambda

# Check odd or even numbers in the list

numbers = [7, 18, 10, 45, 8, 93, 33]

result = list(map(lambda x :  x % 2 == 0, numbers))

print(result)
```

```python
# 2) filter():

#             The filter() function => filters "elements" (from an iterable)
#             based on a 'function'
#             that "returns" either (True or False).

#             The elements that return True are included in the result.


# Syntax:

#               filter(function, iterable)
```

```python
# Examples :  Using filter

# Voting eligibility criteriia

ages = [18, 45, 7, 8, 77, 33, 10]

def vote_eligible(ages):

  if ages >= 18:

    return True

  else:

    return False


result = filter(vote_eligible, ages)

print(list(result))
```

```python
# Examples :  Using filter  with lambda

# Voting eligibility criteria


ages = [18, 45, 7, 8, 77, 33, 10]

vote_eligible = list(filter(lambda x : x >= 18, ages))

print(vote_eligible)
```

```python
# 3) reduce():

#             The reduce() function (from the functools module)
#             It applies a (function of two arguments) cumulatively to the "elements of an iterable",
#             "reducing" it to a (single value).

#             It is often used for operations like (summing or multiplying elements).

# # Syntax:

#              reduce(function, iterable)
```

```python
# Examples :  with only reduce()

from functools import reduce

# product of numbers in the given list

# function
def product(x, y):

  return x * y


lists = [ 7, 2, 4, 3]   # iterable

result = reduce(product, lists)

print("Product of Numbers in the List = ", result)
```

```python
# Examples :  reduce() with lambda function

from functools import reduce

# product of numbers in the given list

lists = [ 7, 2, 4, 3]   # iterable

product = reduce(lambda x, y : x * y, lists)

print(product)
```
