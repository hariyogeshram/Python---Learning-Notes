```python
#                                                                       Operator Overloding

# Definition :

#                   -> It allows operators like( +, -, *), etc., to perform different operations.
#                   -> depending on the "types of objects" they are applied to.

#                   -> This means you can define how "operators" (behave for user-defined types) =>  (like classes).
```

```python
# Where it is used ?

#                     1) Custom Classes              :   To define custom behavior for operators in user-defined classes.

#                     2) Mathematical Operations     :   Overloading operators like +, * for objects such as vectors, matrices, etc.

#                     3) String Operations           :   Overloading + for concatenation or other string manipulations in custom string classes.
```

```python
# Why is it useful?

#                      1) Increased Readability     :    Allows more intuitive and natural expressions (e.g., using + to add complex numbers).

#                      2) Flexibility               :    Custom objects can behave like built-in types.

#                      3) Reusability               :    Allows operators to work with different types of objects without changing the operator logic.
```

```python
# Simple Examples:

# integer

a = 10

b = 20

c = a + b

print("Sum = ", c)
```

```python
# Without using Operator Overloading Examples - 1 ( + )

class Student():

  def __init__(self, mark1, mark2):

    self.mark1 = mark1
    self.mark2 = mark2



obj_1 = Student(18, 45)

obj_2 = Student(8, 7)

result = obj_1 + obj_2

print("Toatl Result = ", result)

```

```python
# With using Operator Overloading Examples - 1 ( + )

class Student():

  def __init__(self, mark1, mark2):

    self.mark1 = mark1
    self.mark2 = mark2


  # Operator Overloading

  def __add__(self, temp):     # self => obj_1   ,   temp => pointing "Obj_2"

    student_1 = self.mark1 + self.mark2
    student_2 = temp.mark1 + temp.mark2

    # calling the class ie : like object creation

    total = Student(student_1, student_2)

    return total



obj_1 = Student(18, 45)

obj_2 = Student(8, 7)


total = obj_1 + obj_2

print("Total for obj_1 = ", total.mark1)
print('')
print("Total for obj_2 = ", total.mark2)
```

```python
# Example - 2 :

class Point:

    def __init__(self, x, y):

        self.x = x
        self.y = y

    # Overloading the '+' operator

    def __add__(self, other):

        return Point(self.x + other.x, self.y + other.y)

    # Displaying point

    def __str__(self):

        return f"Point({self.x}, {self.y})"



# Creating two Point objects
p1 = Point(2, 3)
p2 = Point(4, 1)

# Adding the two points using the overloaded + operator
result = p1 + p2

print(result.x)

print(result.y)

print(result)
```
