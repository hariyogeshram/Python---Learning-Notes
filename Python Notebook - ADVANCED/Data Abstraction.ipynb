```python
# Data Abstraction:

#  Hiding the Internal(functionality) details
#  only showing the essential features to the User


# In Python, abstraction is implemented using "abstract classes" and "abstract methods"
# through the abc (Abstract Base Class) module.
```

```python
# Import the Module

from abc import ABC, abstractmethod


# Abstract class

class Shape(ABC):

    # Abstract method

    @abstractmethod
    def area(self):

        pass

    # Abstract method

    @abstractmethod
    def perimeter(self):

        pass
```

```python
# Concrete Classes (Implementing Abstract Methods)


class Rectangle(Shape):

    def __init__(self, width, height):    # Constructor Initializing

        self.width = width
        self.height = height

    # Implementing the abstract method 'area'
    def area(self):

        return self.width * self.height         #  Area of a Rectangle = (length * breadth)

    # Implementing the abstract method 'perimeter'
    def perimeter(self):

        return 2 * (self.width + self.height)   #  Perimeter of a Rectangle = 2 * (length + breadth)

class Circle(Shape):

    def __init__(self, radius):   # constructor

        self.radius = radius

    # Implementing the abstract method 'area'
    def area(self):

        return 3.1416 * self.radius ** 2    #  Area of a Circle = (pi * radius * radius)

    # Implementing the abstract method 'perimeter'
    def perimeter(self):

        return 2 * 3.1416 * self.radius    #  Perimeter of a Circle = (2 * pi * radius)
```

```python
# Creating objects of Rectangle and Circle

rect = Rectangle(5, 10)
circ = Circle(7)

print("Area of a Rectangle = ", rect.area())
print("Perimeter of a Rectangle = ", rect.perimeter())

print('\n')

print("Area of a Circle = ", circ.area())
print("Perimeter of a Circle = ", circ.perimeter())

```
