```python
# Instance Variable
```

```python
# Definition:

#               An instance variable is a variable
#               that is defined inside a method (usually __init__)
#               and is tied to a specific instance of the class.


# Uses:

#               Stores data that is unique to each instance.
#               Can be accessed and modified only by the instance (using self).
#               Helpful in scenarios where different instances of a class need to maintain separate states.
```

```python
# Examples - ( Instance Variable )

class Employee:

    def __init__(self, name, salary):

        self.name = name        # instance variable
        self.salary = salary    # instance variable

# Creating instances of the class
emp1 = Employee("Hari", 50000)
emp2 = Employee("Ram", 60000)

print(emp1.name, emp1.salary)
print(emp2.name, emp2.salary)
```

```python
# static or class variable
```

```python
# Definition:

#               A static or class variable is shared across all instances of the class.
#               It is defined inside the class but outside any instance methods.
#               The same value is maintained for all instances.


# Uses:


#               Stores data that is shared across all instances.
#               Useful for maintaining common data or constants,
#               such as counting the number of objects created or a configuration shared by all objects.
```

```python
# Examples - [ static or class variable ]

class Employee:

    company = "Cognizant"  # static (class) variable

    def __init__(self, name, salary):

        self.name = name    # instance variable
        self.salary = salary # instance variable


# Creating instances of the class:

emp1 = Employee("Hari", 50000)
emp2 = Employee("Ram", 60000)

print(emp1.name, emp1.salary, emp1.company)
print(emp2.name, emp2.salary, emp2.company)

# Changing the static variable:

Employee.company = "Accenture"

print("\n")

print(emp1.name, emp1.salary, emp1.company)
print(emp2.name, emp2.salary, emp2.company)
```
