```python
# __name__:

# The __name__ variable holds the name of the module currently being executed.

# When a Python script is run directly, the value of __name__ is set to "__main__".

# If a script is imported as a module in another script, __name__ is set to the module's name instead of "__main__".
```

```python
# __main__:

# __main__ refers to the environment where the top-level script is being executed.

# It allows code to be executed when a script is run directly but prevents it from running if the script is imported as a module.
```
