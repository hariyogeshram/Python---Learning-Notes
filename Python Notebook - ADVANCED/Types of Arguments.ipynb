```python
#                                                                   1) Position Argument:

# It will be Assigning (Based on positions)

# Examples : 1

def person(name, age):

  print("Name : ", name)
  print("Age : ", age)

person("Hari", 24)
```

```python
#                                                                        2) keyword Argument

# Example - 1:

def person(name, age):

  print("Name : ", name)
  print("Age : ", age)

person(age = 24, name = "Yogesh")
```

```python
#                                                                            4) Variable-length Arguments

#  Examples :

def add(x, *y):

  print("x = ", x)
  print("y = ", y)

  z = x + y

  print("Sum = ", z)


add(10,20,30,40)
```

```python
# Fix the Above code error


def add(x, *y):

  print("x = ", x)
  print("y = ", y)


  z = x

  for i in y:

    z = z + i

  print("Sum = ", z)


add(10,20,30,40)
```

```python
# Keyword aarguments (**kwargs)

# Examples:

def person(name, **values):

  print("Name : ", name)

  print("Values : ", values.items())


  for i, j in values.items():

    print(i," : ", j)


# Call the Function

person("Hari", age=24, city="Chennai", mailId="hariyogeshram882@gmail.com")
```
