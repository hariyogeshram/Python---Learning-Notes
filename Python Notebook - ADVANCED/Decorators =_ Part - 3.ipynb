```python
# Return "Function" as a Value

# Example - 1

def greet(name):

  def hello():

    return "Hello" + " " + name

  print("hello memory number = ", hello)

  return hello


result = greet("Python")

print("result memory number = ", result)

print(result())
```

```python
# Example - 2 :

def upperstring(ref):

  def process():

    data = ref()

    print(ref)

    return data.upper()


  return process



def my_function():

  return "Welcome to PythoN"


print(my_function)

result = upperstring(my_function)

print(result())
```
