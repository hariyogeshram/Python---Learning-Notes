```python
#                                                                             Inheritance

#  Definition:

#               -> It allows a class
#               -> To Inherit  "Attributes" & "Methods" from another class

#  Main Uses:

#              CODE RE-USABILITY
```

```python
# Types of Inheritance :

#                         1) Single
#                         2) Multiple
#                         3) Multi-Level
#                         4) Hybrid
#                         5) Hierarichal
```

```python
# 1) Single-Inheritance

#  Definition :  It occurs when a class inherits from (only one parent class).
#                The 'child class' gains access to the "properties and methods" of the (single parent class).

# Examples:

class Person():                # Base class,   super class,   parent class

  def person_details(self):

    name = "Yogesh"
    age = 24
    gender = "Male"

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)



class Update_details(Person):    # Derived class,  Sub class,  child class

  def update_person_details(self):

    city = "Chennai"
    mail = "hariyogeshram882@gmail.com"

    print("City : ", city)
    print("Mail-ID : ", mail)



# Object creation

obj = Update_details()

obj.person_details()    # method inherits from base class

obj.update_person_details()   # Method in derived class
```

```python
# 2) Multi-Level Inheritance

#  Definition :                 It occurs when a class inherits from another class, and then another class inherits from that derived class.

# Examples :

class Person():     # Parent class

  def person_details(self, name, age, gender):

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)


class Company(Person):    # child class of Person

  def company_details(self, company_name, location):

    print("Company Name : ", company_name)
    print("HeadQuarters : ", location)


class Other(Company):        # child class of Company class

  def other_details(self, salary, designation):

    print("Salary : ", salary)
    print("Designation : ", designation)



# Object creation:

obj = Other()

obj.person_details("Hari", 24, "Male")

print('')

obj.company_details('Cognizant', "Chennai")

print('')

obj.other_details("5 LPA", "Software Engineer Trainee - Java")

print("\n -----------------------------------------------------------------------------------------------------\n")

obj1 = Other()

obj1.person_details("Kavya", 22, "Female")

print('')

obj1.company_details("Eco-System", "Chennai")

print('')

obj1.other_details("6 LPA", "Full Stack Developer")

print('')
```

```python
# 3) Multiple Inheritance

# Definition :                 A class can inherit "attributes" and "methods" from (more than one parent class).

# Examples :

class Person():     # Parent class

  def person_details(self, name, age, gender):

    print("--------   Personal Details   ------------")
    print('')

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)


class Company():    # child class of Person

  def company_details(self, company_name, location):

    print("--------   Company Details   ------------")
    print('')

    print("Company Name : ", company_name)
    print("HeadQuarters : ", location)


class Other():        # child class of Company class

  def other_details(self, salary, designation):

    print("--------   Other Details   ------------")
    print('')

    print("Salary : ", salary)
    print("Designation : ", designation)


# Multiple Inheritance :

class Employee_Details(Person, Company, Other):

  print("********************************************************************    Employee Details   ****************************************************** ", end="\n\n")

# Object creation:

obj = Employee_Details()

obj.person_details("Hari", 24, "Male")

print('')

obj.company_details('Cognizant', "Chennai")

print('')

obj.other_details("5 LPA", "Software Engineer Trainee - Java")

print("\n --------------------------------------------------------------------------------------------------------------------------------------------------------------\n")

obj1 = Employee_Details()

obj1.person_details("Kavya", 22, "Female")

print('')

obj1.company_details("Eco-System", "Chennai")

print('')

obj1.other_details("6 LPA", "Full Stack Developer")

print('')
```

```python
# 3) Multiple Inheritance  => using super() keyword

# Definition :                 A class can inherit "attributes" and "methods" from (more than one parent class).

# Examples :

class Person():     # Parent class

  def person_details(self, name, age, gender):

    print("--------   Personal Details   ------------")
    print('')

    print("Name : ", name)
    print("Age : ", age)
    print("Gender : ", gender)


class Company():    # child class of Person

  def company_details(self, company_name, location):

    print("--------   Company Details   ------------")
    print('')

    print("Company Name : ", company_name)
    print("HeadQuarters : ", location)


class Other():        # child class of Company class

  def other_details(self, salary, designation):

    print("--------   Other Details   ------------")
    print('')

    print("Salary : ", salary)
    print("Designation : ", designation)


# Multiple Inheritance :

class Employee_Details(Person, Company, Other):

  def show_employee_details(self):

     print("********************************************************************    Employee Details   ****************************************************** ", end="\n\n")

     super().person_details("Hari", 24, "Male")      # Override the method from "Person" class by using super()

     super().company_details('Cognizant', "Chennai")   # Override the method from "Company" class by using super()

     super().other_details("5 LPA", "Software Engineer Trainee - Java")   # Override the method from "Other" class by using super()



# Object creation:

obj = Employee_Details()

obj.show_employee_details()
```

```python
#  4) Hierarichal Inheritance

#  Definition :                  It occurs when 'multiple child classes' inherit (from a single parent class)

# Examples:

# Parent class
class Person:

    def person_info(self):

        return "Person: General Information"

# Child class 1 (inherits from Person)
class Student(Person):

    def student_info(self):

        return "Role: Student"

# Child class 2 (inherits from Person)
class Teacher(Person):

    def teacher_info(self):

        return "Role: Teacher"

# Create objects of Student and Teacher
student = Student()
teacher = Teacher()

print(student.person_info())  # Output: Person: General Information
print(student.student_info()) # Output: Role: Student
print('\n-------------------------------------\n')
print(teacher.person_info())  # Output: Person: General Information
print(teacher.teacher_info()) # Output: Role: Teacher
```

```python
#  5) Hybrid Inheritance

#  Definition :                   'combination' of "two or more types of inheritance" in a (single program).
#                                 It involves (multiple inheritance types) like (single, multiple, hierarchical, or multilevel).
#                                 The goal is to utilize the "strengths of different inheritance mechanisms" to 'create' a (flexible and robust structure).

# Examples:

# Base Parent class
class Person:

    def person_info(self):

        return "This is a person"

# Single inheritance: Student inherits from Person
class Student(Person):

    def student_info(self):

        return "This is a student"

# Single inheritance: Teacher inherits from Person
class Teacher(Person):

    def teacher_info(self):

        return "This is a teacher"

# Multiple inheritance: Assistant inherits from both Student and Teacher
class Assistant(Student, Teacher):

    def assistant_info(self):

        return "This is a teaching assistant"

# Create objects for each class
assistant = Assistant()

# Access methods from both parents
print(assistant.person_info())   # Output: This is a person
print(assistant.student_info())  # Output: This is a student
print(assistant.teacher_info())  # Output: This is a teacher
print(assistant.assistant_info())# Output: This is a teaching assistant
```

```python
# Same -> "Method Name" in 'Inheritance'

# Example - 1:

class A():

  def greet(self):

    return "Hi..This is from 'A' Class"

class B():

  def greet(self):

    return "Hello..This is from 'B' Class"

class C(A, B):

  def greet(self):

    return "Welcome to 'C' Class"


# Obj creation

obj = C()

obj.greet()
```

```python
# Constructor in Inheritance

# Example - 1 :

# Base class

class Teacher():

  # constructor

  def __init__(self, s):

    self.subject = s

    print ("I am a Teacher")


# Derived class

class Student(Teacher):

  def details(self):

    print("I am a Student")

    print("My favourite Subject : ", self.subject)



# Object Creation

obj = Student("Physics")      # output : I am a Teacher

obj.details()
```

```python
# super()

# Purpose : super() is used to 'give access' to "methods and properties" of a (parent or sibling class) in a 'child class'.

# Common Usage: It is typically used (when you want to override a method) in a "subclass"
#               but still need to call the parent class’s version of that method to maintain functionality.



#  Examples :

class Person():    # parent class

  def __init__(self, name, age):

    self.name = name
    self.age = age

  def display_details(self):

    print(f" Name : {self.name}\n Age : {self.age}")


# child class

class Student(Person):

  def __init__(self, name, age, student_id, designation, city):

    # Call the parent class's constructor using super()

    super().__init__(name, age)

    self.student_id = student_id
    self.designation = designation
    self.city = city


  # Override the display_details method

  def display_details(self):


    # Call the 'parent class's' ("display_details") method using super()

    super().display_details()    # Access "Person's" (display_details) method

    #        or

    # Person.display_details(self)   # If you give directly parent class => you must give parameters : (self)

    print(f" Student ID : {self.student_id}\n Designation : {self.designation}\n City : {self.city}")



# (Creating an object) of 'Student' class

obj = Student("Hari", 24, "Stu_ID_45", "Full_Stack_Developer", "Chennai")

obj.display_details()
```

```python
# Example - 2 (super() keyword)

class Shape():

  def __init__(self, length, breadth):

    self.length = length
    self.breadth = breadth

    self.result = self.length * self.breadth


class Square(Shape):     # Area of square = side * side

  def __init__(self, length):

    super().__init__(length, length)

    print("Area of a Square = ", self.result)



class Rectangle(Shape):     # Area of Rectangle = Length * breadth

  def __init__(self, length, breadth):

    super().__init__(length, breadth)

    print("Area of a Rectangle = ", self.result)



class Cube(Shape):     # Area of Cube = Length * breadth * height

  def __init__(self, length, breadth, height):

    super().__init__(length, breadth)

    self.height = height

    print("Area of a Cube = ", self.result * self.height)




# Object creation

s = Square(4)

r = Rectangle(5, 3)

c = Cube(4, 2, 5)
```
