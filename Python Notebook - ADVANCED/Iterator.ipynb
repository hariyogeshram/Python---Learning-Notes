```python
# iterator is an object
# that allows you to "traverse through all the elements" in a "collection" (like lists, tuples, or dictionaries) one element at a time,
# without needing to know the underlying structure.


# An iterator implements two essential methods:

# __iter__(): Returns the iterator object itself.
# __next__(): Returns the next item from the collection.

# When (there are no more items to return) =>  it raises a "StopIteration" exception.
```

```python
# Normal Iterating Examples

lst = ["hai", "hello", 45, "morning", 180.1, True]

# Using 'for' loop

for i in lst:

  print(i)
```

```python
lst = ["hai", "hello", 45, "morning", 180.1, True]

# Using 'while' loop

i = 0

while i < len(lst):

  print(lst[i])

  i+=1
```

```python
# Original Iterator Examples:

lists = [10, 18, 45, 7, 93, 33, 1]



# iter() => Get an iterator Object

result = iter(lists)

# print(result)        # <list_iterator object at 0x7f64261ce920>



# Using next() => To Access "Elements (one by one)"

print(next(result))    # 10

print(next(result))    # 18

print(next(result))    # 45

print(next(result))    # 7

print(next(result))    # 93

print(next(result))    # 33

print(next(result))    # 1


# print(next(result))    # Raise an "StopIteration" Exception
```

```python
# [ Creating a Custom Iterator ]

# You can also create your own custom iterator by defining the __iter__() and __next__() methods in a class.
```

```python
# Examples:

# Example: Custom Iterator that generates squares of numbers

class SquareIterator:

    def __init__(self, max_value):

        self.max_value = max_value

        self.current = 1

    def __iter__(self):    # Returns an iterator object

        return self

    def __next__(self):

        if self.current <= self.max_value:

            result = self.current ** 2

            self.current += 1

            return result


        else:

            raise StopIteration

# Using the custom iterator:

squares = SquareIterator(5)

for number in squares:

    print(number)
```
