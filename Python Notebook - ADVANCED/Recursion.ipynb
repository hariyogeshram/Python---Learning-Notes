```python
# Recursion

#   The function that calls itself to solve a problem.
```

```python
# check the recursion limit

import sys

print("Recursion Default Limit = ", sys.getrecursionlimit())
```

```python
# Recursion Example - 1 :   [ Factorial Number]

def factorial(n):

  if n==0 or n==1:

    return 1

  else:

    return n*factorial(n-1)


# function call

n = int(input("Enter the Number : "))

result = factorial(n)

print(f"Factorial of {n} = {result}.")
```

```python
# Recursion Example - 2 :   [ Fibonacci series]

def fibonacci_series(n):

  if n==0:

    return 0

  elif n==1:

    return 1

  else:

    return fibonacci_series(n-1) + fibonacci_series(n-2)


# Function call

n = int(input("Enter how many integer you want to print : "))

result = fibonacci_series(n)

print(f"Fibonacci series of {n} intergers = {result}.")
```
