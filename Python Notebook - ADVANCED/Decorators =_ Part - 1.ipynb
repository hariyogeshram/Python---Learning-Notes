```python
#  Definition:


# decorator is a "design pattern"
# that allows you to "modify or extend" the "behavior of functions or methods" (without permanently modifying them).
# A decorator is itself a 'function' that takes 'another function' (or method) as an "argument" and 'returns' a "new function"
# that typically 'wraps' the "original one".
```

```python
# Nested Function
# Return Function from another function
# Function as a Reference

# Examples :

def outer():

  message_1 = "Happy"

  def inner():

    message_2 = "Learning"

    result = message_1 +" "+message_2

    return result


  return inner



# object creation for the method declaration

obj = outer()

print(obj())
```

```python
def simple():

  print("Hello!..")

print(simple)

simple()
```

```python
def outer(x):

  def inner(y):

    return x + y

  return inner


add = outer(5)

result = add(8)

print(result)
```
